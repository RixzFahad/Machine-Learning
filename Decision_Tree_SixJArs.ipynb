*Q) Did we use gradient descent in Decision Tree?*
a) No

**Q) Why not?**
Decision Trees do not learn by optimizing weights
They split data based on rules (questions), not by minimizing a loss function using gradients

***Q) If not, then how do we find the best or maximum depth of a Decision Tree?***
a) Using splitting criteria and validation techniques

→ Algorithm finds best splits using:
→ entropy / gini index (to choose best question at each node)

→ Depth is controlled by:
→ max_depth (hyperparameter)
→ min_samples_split
→ min_samples_leaf

→ Best depth is found using:
→ cross-validation
→ checking train vs test accuracy

→ Key idea:
→ too deep tree → overfitting
→ too shallow tree → underfitting

→ So we choose depth where model generalizes best
