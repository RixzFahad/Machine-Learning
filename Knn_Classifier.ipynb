# ***Knn Classifier***

*KNN (K-Nearest Neighbors) is a supervised learning algorithm used for classification and regression*

**It works by finding the ‘K’ closest data points (neighbors) to a new input based on distance (like Euclidean distance)**

***For classification: it assigns the class which is most common among the neighbors***
***For regression: it takes the average of the neighbors’ values***

**It is a non-parametric and lazy learning algorithm (no training phase, stores all data)**

*Example: if K = 3 and among nearest neighbors → [Yes, Yes, No] → output = Yes*

_**q) What type of task is the above problem?**_ -> Supervised Learning (SL)

_**Q) In SL tasks, what type of problem is this?**_ -> Classification

_**q) Why is it classification?**_ -> We predict categorical values (e.g., yes or no).

_**q) Can I apply logistic regression?**_ -> No.

_**q) Why not?**_ -> Because there is no linear decision boundary.

_**q) So, if I still want to predict on this classification data, what do we need to do?**_

***Question1-*** **How can We Measure The Nearness?**

***Answer*** **By using the distance matrix - Euclidean distance\**

```python
# what type of task is above problem -> SL?
# In SL task what type of problem -> classification
# why it is classification -> we predict categorical values -> yes or no
# can i apply logistic regression -> No -> why -> because there is no linear decision boundary
# so still i want to predict on this classification data what we need to do?
#
# euclidean distance calculates distance between 2 data points (x1, y1), (x2, y2)
# (2, 5) (6, 8)
# = sqrt((x2 - x1)^2 + (y2 - y1)^2)
# = sqrt((6 - 2)^2 + (8 - 5)^2)
# = sqrt(4^2 + 3^2)
# = sqrt(16 + 9)
# = sqrt(25) = 5
```

_***Steps for applying KNN algorithm:***_

1.  _**We need to get the "k" value from a dataset (e.g., k = 3)**_
2.  _**Compute the distance between all the data points (e.g., for 3-NN)**_
3.  _**After finding the distance, choose the nearest values**_
4.  _**Predict based on the majority nearest point**_

_**Optimal 'k' value:**_

*   _**Q) What is the maximum value that "k" can take?**_
    *   _**A) It can be any positive integer.**_

*   _**Q) How to find the optimal "k" value or how to find the best "K" value?**_

_**In K-NN, did we get the "k" value automatically?**_ -> No

_**The ML algorithm model whenever we giving the values or (we feed the values) is known as "hyperparameter model"**_

_**The ML algorithm automatically finds its parameters value is known as "parametric model" -> linear regression and logistic regression.**_

_**Q) How do I know that k is 1, 2, 3, 4, 5, 6, 7, 8, 9...?**_
_**A) To find the hyperparameter (k-value), we need to use a "hit and trial" method.**_

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Generate a synthetic dataset
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List to store accuracy for different k values
accuracies = []
k_values = range(1, 21) # Test k from 1 to 20

print("Evaluating K-Nearest Neighbors for different 'k' values...")
for k in k_values:
    # 3. Train a KNN classifier for each k
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # 4. Predict on the test set
    y_pred = knn.predict(X_test)

    # 5. Calculate and store the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"  k = {k}: Accuracy = {accuracy:.4f}")

# 6. Plot the results to visualize the optimal k
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='b')
plt.title('KNN Accuracy vs. Number of Neighbors (k)')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.grid(True);
plt.show()

# Find the optimal k value
optimal_k = k_values[np.argmax(accuracies)]
max_accuracy = np.max(accuracies)
print(f"\nOptimal 'k' value: {optimal_k} with an accuracy of {max_accuracy:.4f}")
```

**Answer to Question 1: Explain the K-Nearest Neighbors (KNN) algorithm. How does it work for both classification and regression tasks?**

KNN is a non-parametric, lazy learning algorithm. For **classification**, it assigns a data point to the class that is most common among its 'k' nearest neighbors. For **regression**, it predicts the average (or median) value of its 'k' nearest neighbors.

**Answer to Question 2: What is the primary difference between a supervised and unsupervised learning algorithm? Provide an example of each.**

**Supervised learning** uses labeled data (input-output pairs) to train a model to make predictions (e.g., classification, regression - like KNN or Linear Regression). **Unsupervised learning** works with unlabeled data to find patterns or structures without explicit guidance (e.g., clustering - like K-Means).

**Answer to Question 3: KNN is often referred to as a 'lazy learning' algorithm. What does this mean, and what are its implications?**

'Lazy learning' means KNN does not build a model during the training phase; it simply stores the entire dataset. Prediction occurs only when a new query point is introduced, calculating distances to all training points. Implications include high computational cost at prediction time and a lack of an explicit model to interpret.

**Answer to Question 4: How do you determine the 'nearest' neighbors in KNN? Describe the most commonly used distance metric and its formula.**

'Nearest' neighbors are determined by a **distance metric**. The most commonly used is **Euclidean distance**. For two points (x1, y1) and (x2, y2), its formula is `sqrt((x2 - x1)^2 + (y2 - y1)^2)`. More generally, for N dimensions, it's the square root of the sum of the squared differences of corresponding coordinates.

**Answer to Question 5: What is the role of the 'k' parameter in KNN? How does the choice of 'k' affect the model's performance and decision boundary?**

'k' is the number of nearest neighbors considered for making a prediction. A **small 'k'** makes the model more sensitive to noise (high variance, low bias, complex decision boundary). A **large 'k'** smooths out decision boundaries, making it less sensitive to noise but potentially missing local patterns (low variance, high bias, simpler decision boundary).

**Answer to Question 6: Explain the difference between a 'parametric' and 'non-parametric' model. Where does KNN fall, and why?**

**Parametric models** assume a fixed number of parameters and a specific functional form for the data (e.g., Linear Regression). **Non-parametric models** do not assume a fixed form and can adapt their structure based on the data (e.g., KNN, Decision Trees). KNN is **non-parametric** because it doesn't learn a specific function; it relies directly on the training data points for each prediction.

**Answer to Question 7: How would you go about finding an optimal 'k' value for a KNN model in a real-world scenario? Discuss potential methods beyond the 'hit and trial' shown in this notebook.**

Beyond 'hit and trial', optimal 'k' can be found using **cross-validation** (e.g., k-fold cross-validation) combined with **grid search** or **randomized search**. These methods systematically evaluate various 'k' values and select the one that performs best on a validation set, minimizing bias from a single train-test split.

**Answer to Question 8: What are some of the advantages and disadvantages of using KNN? When would you choose KNN over other classification algorithms like Logistic Regression?**

**Advantages:** Simple, intuitive, no training phase, effective for multi-class problems. **Disadvantages:** Computationally expensive during prediction, sensitive to irrelevant features, sensitive to data scale, affected by the curse of dimensionality. Choose KNN when decision boundaries are highly non-linear, the dataset is small to medium-sized, and local patterns are important.

**Answer to Question 9: What is the curse of dimensionality, and how does it relate to KNN?**

**The curse of dimensionality** refers to various problems arising from high-dimensional data, such as data sparsity and distances becoming less meaningful. In KNN, as the number of dimensions increases, the distance between any two points tends to become more uniform, making it harder to distinguish true 'nearest' neighbors and leading to reduced model performance.
