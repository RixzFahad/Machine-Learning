# ***Machine Learning Introduction***

* Machine learning is a part of artificial intelligence
* It allows systems to learn from data instead of coding rules
* Used for prediction and decision making

**Learns patterns automatically from data**

# ***Types of Machine Learning***

* Supervised learning → works with labeled data
* Unsupervised learning → finds hidden patterns
* Reinforcement learning → learning using rewards

**Three main approaches of learning**

# ***Basic Terminology***

* Dataset → collection of data
* Features → input variables
* Label → output variable
* Model → learns patterns
* Training → learning phase
* Testing → evaluation phase

# ***Machine Learning Workflow***

* Data collection
* Data cleaning
* Model training
* Evaluation
* Deployment

**Complete machine learning pipeline**

# ***Model Concept***

* Model is a function mapping input to output
* Learns relationship between variables
* Improves with more data

**Example → y = mx + c**

# ***Algorithms***

* Linear regression → predicts continuous values
* Logistic regression → used for classification
* Decision tree → rule-based structure
* Random forest → multiple trees combined
* KNN → nearest neighbor method
* SVM → finds best boundary

# ***Overfitting vs Underfitting***

* Overfitting → memorizes data, fails on new data
* Underfitting → too simple, cannot learn patterns
* Goal → balanced model

**Good model generalizes well**

# ***Evaluation Metrics***

* Accuracy → correct predictions
* Precision → correct positive predictions
* Recall → actual positives identified
* F1 score → balance of precision and recall

**Used to measure performance**

# ***Applications***

* Healthcare → disease prediction
* Finance → fraud detection
* E-commerce → recommendation systems
* Social media → feed ranking
* Self driving → object detection

```python
# Simple Example

import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])

model = LinearRegression()
model.fit(X, y)

pred = model.predict([[5]])
print(pred)

# Model learned linear relationship
```
